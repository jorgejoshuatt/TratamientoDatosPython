```python
#Cargar el DF de categorias
df_cat=pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/categorias.csv')
df_cat
```

```python
#Visualizar los valores unicos de cada columna categorica
#unique()
print('Limpieza ',df['limpieza'].unique())
print('Seguridad ',df['seguridad'].unique())
print('Satisfaccion ',df['satisfaccion'].unique())
```

```python
#Localizar los valores invalidos en la columna limpieza
df_limpieza=set(df['limpieza']).difference(df_cat['limpieza'])
df_limpieza
```

```python
df_lim_r=df['limpieza'].isin(df_limpieza)
df[df_lim_r]
```

```python
#Contar valores que tienen valores invalidos en limpieza
df[df_lim_r].count()
```

```python
#Contar valores que tienen valores validos en limpieza
df[~df_lim_r].count()
```

```python
#Reemplazar valores en una columna
df['limpieza']=df['limpieza'].replace({'Inaceptable':"Sucio",'Medio':"Promedio"})
set(df['limpieza']).difference(df_cat['limpieza'])
```

```python
#Validar columnas region, dest_size
print('Region',df['region'].unique())
print('Dest Size',df['dest_size'].unique())
```

```python
#Transformar datos
#Pasar a minusculas
df['region']=df['region'].str.lower()

#Reemplazar valores
df['region']=df['region'].replace({'eur':'europa'})

#Quitar espacios en blanco
df['dest_size']=df['dest_size'].str.strip()

#Usar expresiones regulares para reemplazo
df['dest_size']=df['dest_size'].str.replace("grande","large",regex=True, flags=re.IGNORECASE)
```

```python
#Validar dia
print('Dia',df['dia'].unique())
```

```python
#Generar valores validos para dia
reemplazo={'Thursday':'jueves','Sunday':'domingo','\u200b\u200bmartes':'martes', 'Friday':'viernes'}
df['dia']=df['dia'].replace(reemplazo)
df['dia']=df['dia'].str.lower()
#Validar dia
print('Dia',df['dia'].unique())
```

```python
#Generar una columna categorica
#espera_tipo (corto, medio, larga)

espera_rango=[0,60, 180, np.inf]
espera_etiqueta=['corto', 'medio', 'largo']
df['espera_tipo']=pd.cut(df['espera_min'],bins=espera_rango, labels=espera_etiqueta)
df[['espera_min','espera_tipo']]
```

```python
#Generar columna categorica
#Tipo dia de la semana
mappings = {'lunes':'weekday', 'martes':'weekday', 'miércoles': 'weekday', 
            'jueves': 'weekday', 'viernes': 'weekday', 
            'sábado': 'weekend', 'domingo': 'weekend'}
df['tipo_dia_semana']=df['dia'].replace(mappings)
df[['dia','tipo_dia_semana']]
```

```python
df['tipo_dia_semana'].describe()
```

```python
#Convertir una columna en tipo category
df['tipo_dia_semana']=df['tipo_dia_semana'].astype("category")
df.info()
```

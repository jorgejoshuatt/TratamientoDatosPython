```python
import pandas as pd
import numpy as np
```

```python
#Leer archivo csv y mostrar la información
df_titanic=pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv')
df_titanic.info()
```

```python
#Mostrar la tabla del DF
df_titanic
```

```python
#Visualizar el número de pasajeros sin el dato edad y la tabla
df_edad_nan=df_titanic[pd.isnull(df_titanic['age'])]['age'].isna().sum()
print("Numero de pasajeros sin el dato edad: ", df_edad_nan)
df_edad_nan=df_titanic[pd.isnull(df_titanic['age'])]['age'].isna()
df_titanic[df_titanic.index.isin(df_edad_nan.index)]
```

```python
# 1. Si la edad es null (na) llenar el campo con la media de la edad (usar el metodo fillna())
df_titanic['age'].fillna((df_titanic['age'].mean()),inplace=True)
```

```python
#Visualizar los cambios en la tabla después de rellenar con fillna(), se puede observar que la media es 29.881135
df_titanic[df_titanic.index.isin(df_edad_nan.index)]
```

```python
#Visualizar el número de pasajeros sin el dato edad, corroboramos que ya nigún dato en edad está vacío
df_edad_nan=df_titanic[pd.isnull(df_titanic['age'])]['age'].isna().sum()
print("Numero de pasajeros sin el dato edad: ", df_edad_nan)
```

```python
#Visualizar el número de pasajeros sin el dato embarked y la tabla
df_embarked_nan=df_titanic[pd.isnull(df_titanic['embarked'])]['embarked'].isna().sum()
print("Numero de pasajeros sin el dato embarked: ", df_embarked_nan)
df_embarked_nan=df_titanic[pd.isnull(df_titanic['embarked'])]['embarked'].isna()
df_titanic[df_titanic.index.isin(df_embarked_nan.index)]
```

```python
# 2. Si embarqued es na llenar con S
df_titanic['embarked'].fillna("S",inplace=True)
```

```python
#Visualizar los cambios en la tabla después de rellenar con fillna()
df_titanic[df_titanic.index.isin(df_embarked_nan.index)]
```

```python
#Visualizar el número de pasajeros sin el dato embarked y la tabla, podemos corroborar que ningún dato en embarked está vacío
df_embarked_nan=df_titanic[pd.isnull(df_titanic['embarked'])]['embarked'].isna().sum()
print("Numero de pasajeros sin el dato embarked: ", df_embarked_nan)
```

```python
# 3. Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }
df_titanic['sex']=df_titanic['sex'].replace({'female':1,'male':0})
df_titanic['sex']
```

```python
# 4. Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }
df_titanic['embarked']=df_titanic['embarked'].replace({'S':0,'C':1,'Q':2})
df_titanic['embarked']
```

```python
# 5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas
child_rango=[0,16, np.inf]
child_etiqueta=[1, 0]
df_titanic['child']=pd.cut(df_titanic['age'],bins=child_rango, labels=child_etiqueta)
df_titanic[['age','child']]
```

```python
# 6. Eliminar las columnas ticket, cabin y name
df_titanic.drop(['ticket','cabin','sibsp'],axis=1,inplace=True)
df_titanic
```

```python
# 7. Validar que no existan registros duplicados
df_dup_titanic=df_titanic.duplicated()
df_dup_titanic
```

```python
#Mostrar valores duplicados, se puede observar que no hay
df_titanic[df_dup_titanic]
```

```python
# 8. Listar los valores unicos del campo sex
print('Sexo ',df_titanic['sex'].unique())
```

```python
# 9. Quitar los espacios vacios de los campos que no sean numericos
df_titanic['home.dest']=df_titanic['home.dest'].str.strip()
```

```python
# 10. Cambiar el nombre de las columnas de inglés a español
df_titanic=pd.DataFrame(df_titanic)
df_titanic.columns=['tboleto', 'sobrevivio', 'nombre', 'sexo', 'edad', 'npadres/hijos', 'costoboleto', 'embarcacion', 'numbote', 'numcuerpo', 'domicilio/destino', 'nino']
df_titanic
```

```python
# 11. Mostrar los estadisticos básicos de la columna edad
df_titanic['edad'].describe()
```

```python
# 12. En un df nuevo dejar solo las columnas nombre, puerto de embarcación y edad de las mujeres
df_mujeres=df_titanic.loc[df_titanic['sexo']==1,['nombre','embarcacion','edad']]
df_mujeres
```

```python
# 13. Guardar los datos limpios en un archivos json orientado a registros llamado iniciales_titanic_clean.json (sus iniciales)
df_titanic.to_json('jjtm_titanic_clean.json',orient="records")
%ls /content/
```

```python
# 14. Guardar los datos limpios en un archivo de excel con dos hojas una con los sobrevientes y otra con los fallecidos llamado iniciales_titanic_clean.xlsx (sus iniciales)
df_sobrevivio=df_titanic.loc[df_titanic['sobrevivio']==1]
df_fallecio=df_titanic.loc[df_titanic['sobrevivio']==0]
with pd.ExcelWriter('jjtm_titanic_clean.xlsx') as w:
  df_sobrevivio.to_excel(w, sheet_name='Sobrevivientes',index=False)
  df_fallecio.to_excel(w, sheet_name='Fallecidos',index=False)
    
%ls /content/
```
